# Prediction using Decision Tree Algorithm

```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree
```

```python
# Loading iris dataset
iris = datasets.load_iris()

# Forming the iris dataframe
tree_df=pd.DataFrame(iris.data, columns=iris.feature_names)

tree_df.head(10)


```

```python
y=iris.target
print(y)
```

```python
iris.target_names
```

# Creating Tree Classsifier

```python
tree_clf = DecisionTreeClassifier(random_state=62)
model = tree_clf.fit(tree_df, y)
print('Decision Tree Classifer Created')
```

# Textual Representation of tree

```python
text_representation = tree.export_text(tree_clf)
print(text_representation)
```

# Visualizing Decision Tree

```python
fig = plt.figure(figsize=(25,20))
decision_tree = tree.plot_tree(tree_clf, 
                   feature_names=iris.feature_names,  
                   class_names=iris.target_names,
                   filled=True)
```

```python
fig.savefig("decistion_tree.png")
```

```python
#predicting class for new data
predictions=tree_clf.predict([[5.3,4,2.7,2.3]])
print(predictions)

```
